```

import pandas as pd
import ssl
from matplotlib import pyplot as plt
import numpy as np

ssl._create_default_https_context = ssl._create_unverified_context
mouselocal = pd.read_csv('https://hyeshik.qbio.io/binfo/mouselocalization-20210507.txt', sep='\t')

cnts = pd.read_csv('read-counts.txt', sep='\t', comment='#', index_col=0)
cnts['clip_enrichment'] = cnts['CLIP-35L33G.bam'] / cnts['RNA-control.bam']
cnts['rden_change'] = (cnts['RPF-siLin28a.bam'] / cnts['RNA-siLin28a.bam']) / (cnts['RPF-siLuc.bam'] / cnts['RNA-siLuc.bam'])

cnts.index = cnts.index.str.split('.').str[0]

mouselocal.set_index('gene_id', inplace=True)
cnts = cnts.join(mouselocal['type'])

cnts.head()

cnts_filtered = cnts[cnts['RNA-control.bam'] > 200 ]

nucleus = cnts_filtered[cnts_filtered['type'] == 'nucleus']
cytoplasm = cnts_filtered[cnts_filtered['type'] == 'cytoplasm']
integral_membrane = cnts_filtered[cnts_filtered['type'] == 'integral membrane']

fig, ax = plt.subplots(1, 1, figsize=(5, 5))
ax.scatter(np.log2(nucleus['clip_enrichment']), np.log2(nucleus['rden_change']), s=1, color='blue', label = 'nucleus')
ax.scatter(np.log2(cytoplasm['clip_enrichment']), np.log2(cytoplasm['rden_change']), s=1, color='green', label = 'cytoplasm')
ax.scatter(np.log2(integral_membrane['clip_enrichment']), np.log2(integral_membrane['rden_change']), s=1, color='pink', label = 'integral_membrane')
plt.xlabel("LIN28A CLIP enrichment(log2)")
plt.ylabel("Ribosome density change \n on Lin28a knockdown (log2)")
```

```
GO_gaf=GO_gaf[[1,2,4]]
GO_gaf.columns = ['uniprot_id', 'gene_symbol','go_term']
GO_gaf.head()
```

```
GO_gaf.to_csv('GO_gaf.txt', sep='\t', index=False)

#Uniprot.org의 id-mapping을 통해 uniprot id를 ensembl id로 바꾸어서 id mapping table을 다시 구글 드라이브로 옮김.


IDmapping = pd.read_csv('uniprot-compressed_true_download_true_format_tsv-2023.06.07-15.58.22.59.tsv', sep='\t', comment='!')
```

```
IDmapping.head()
```

```
id_mapping_dict = dict(zip(IDmapping['From'], IDmapping['To']))
GO_gaf['Ensembl_id']= GO_gaf['uniprot_id'].map(id_mapping_dict).str.split('.').str[0]

GO_gaf.head()
len(GO_gaf)
```

```
GO_gaf = GO_gaf[~GO_gaf['gene_symbol'].str.startswith('Hist')]

print(GO_gaf)
len(GO_gaf)
```

```
len(GO_gaf)

GO_gaf.to_csv('GO_gaf.txt', sep='\t', index=True, index_label='Row Name')
```

```
#filtering cnts 
cnts.head()
def calculate_rpkm(counts, lengths):
    N = counts.sum()
    L = lengths / 1e3
    C = counts
    return (1e9 * C) / (N * L)

# RNA-siLuc.bam RPKM 계산
rpkm = calculate_rpkm(cnts['RNA-siLuc.bam'], cnts['Length'])

# 15 RPKM 이하의 유전자 제거 (RNA-siLuc.bam)
cnts = cnts[rpkm > 15]

# low read count 제거 
cnts = cnts[(cnts['RPF-siLin28a.bam'] > 80) & (cnts['RPF-siLuc.bam'] > 80)]

cnts = cnts[cnts['RNA-control.bam'] > 150]


# low footprints 제거 
cnts = cnts[(cnts['RNA-siLin28a.bam'] > 30)]

cnts.head()
```

```
import pandas as pd
from scipy.stats import mannwhitneyu
from statsmodels.stats.multitest import multipletests

# cnts에서 ensembl_id와 공통된 유전자만 GO_gaf에서 추출
GO_gaf = GO_gaf[GO_gaf['Ensembl_id'].isin(cnts.index)]
```

```
result_df = pd.DataFrame(columns=['num_genes', 'rden_p_value', 'clip_p_value', 'clip_enrichment_mean', 'rden_change_mean'])

#GO term 하나씩 Mann-Whitney U test 실시
for go_term in GO_gaf['go_term'].unique():
    # 해당 GO term에 속하는 유전자만 선택
    go_genes = GO_gaf[GO_gaf['go_term'] == go_term]

    # GO term에 속하는 유전자들의 rden_change와 속하지 않는 유전자들의 rden_change들에 대한 Mann-Whitney U test
    go_rden_change = cnts.loc[go_genes['Ensembl_id'], 'rden_change']
    non_go_rden_change = cnts.drop(go_genes['Ensembl_id'], errors='ignore')['rden_change']
    
    _, rden_p_value = mannwhitneyu(np.log2(go_rden_change), np.log2(non_go_rden_change))

    # GO term에 속하는 유전자들의 clip_enrichment와 속하지 않는 유전자들의 clip_enrichment에 대한 Mann-Whitney U test
    go_clip_enrichment = cnts.loc[go_genes['Ensembl_id'], 'clip_enrichment']
    non_go_clip_enrichment = cnts.drop(go_genes['Ensembl_id'], errors='ignore')['clip_enrichment']
    
    _, clip_p_value = mannwhitneyu(np.log2(go_clip_enrichment), np.log2(non_go_clip_enrichment))

    # GO term에 속하는 유전자들의 clip_enrichment 값의 평균
    clip_enrichment_mean = (np.log2(go_clip_enrichment)).mean()
    
    # GO term에 속하는 유전자들의 rden_change의 평균
    rden_change_mean = (np.log2(go_rden_change)).mean()
    
    result_df.loc[go_term] = [len(go_genes), rden_p_value, clip_p_value, clip_enrichment_mean, rden_change_mean]

# FDR 계산
_, result_df['rden_FDR'], _, _ = multipletests(result_df['rden_p_value'], method='fdr_bh')
_, result_df['clip_FDR'], _, _ = multipletests(result_df['clip_p_value'], method='fdr_bh')

result_df.columns = ['num_genes', 'rden_p_value', 'clip_p_value', 'clip_enrichment_mean', 'rden_change_mean', 'rden_FDR', 'clip_FDR']

```

```
result_df.to_csv('result_df.txt', sep='\t', index=True, index_label='Row Name')
```

```
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(1, 1, figsize=(15, 5))

# scatter plot의 크기 조절
scale_factor = 0.1

# clip_enrichment_FDR 값에 따라 내림차순으로 정렬 (dot의 배치 조절)
result_df_rden_filtered = result_df_rden_filtered.sort_values('clip_FDR', ascending=False)

# dot의 색상과 size 설정
c = result_df_rden_filtered['clip_FDR'].map(np.log)
s = result_df_rden_filtered['num_genes'] * scale_factor

cmap = plt.cm.get_cmap('YlOrRd')  # 'YlOrRd' 색상 맵 가져오기
cmap_reversed = cmap.reversed()  # 색상 맵을 반대로 설정

scatter = ax.scatter(result_df_rden_filtered['clip_enrichment_mean'],
                     result_df_rden_filtered['rden_change_mean'],
                     c=c,
                     s=s,
                     cmap=cmap_reversed)

# colorbar 추가
cbar = plt.colorbar(scatter)
cbar.set_label('log(clip_enrichment_FDR)')

# x, y 축 범위 및 눈금 설정
ax.set_xlim([-2.0, 5.0])
ax.set_ylim([-2.0, 1.5])
ax.set_xticks(np.arange(-2.0, 5.5, 0.5))
ax.set_yticks(np.arange(-2.0, 1.5, 0.5))

# 그래프 테두리 추가
for spine in ax.spines.values():
    spine.set_visible(True)

# 축 레이블 설정
ax.set_xlabel("Enrichment level of LIN28A-bound CLIP tags (log2)")
ax.set_ylabel("Ribosome density change upon $\it{Lin28a}$ knockdown (log2)")

# 그리드 추가
ax.grid(True, linestyle='--', alpha=0.6)

from adjustText import adjust_text

# 'clip_FDR' 값에 따라 DataFrame을 오름차순으로 정렬하고 상위 10개 row 선택
top_10_terms = result_df_rden_filtered.sort_values('clip_FDR', ascending=True).head(10)

# annotation
annotations = []
for idx, row in top_10_terms.iterrows():
    label = f"{idx} ({int(row['num_genes'])})\nC = {row['clip_FDR']:.2e} R={row['rden_FDR']:.2e}"
    x = (row['clip_enrichment_mean'])
    y = (row['rden_change_mean'])
    annotations.append(ax.text(x, y, label, fontsize=8,
                               bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.5')))

# annotation 위치 최적화
adjust_text(annotations, force_points=0.2, force_text=0.2, 
            expand_points=(1, 1), expand_text=(1, 1), 
            arrowprops=dict(arrowstyle="->", color='black', lw=0.5))

plt.savefig('graph.pdf', format='pdf')
plt.show()

```
